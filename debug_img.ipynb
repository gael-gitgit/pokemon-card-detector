```python
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --- CONFIG ---
path = "data/images/inputs/b0jlcxfsc1ub1.jpg"  # <-- remplace par ton chemin d'image

# --- CHARGEMENT DES IMAGES ---
img_pil = Image.open(path).convert("RGB")
img_cv = cv2.imread(path)

# --- ALIGNEMENT DES FORMATS ---
img_cv_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)

# --- CONVERSION EN TABLEAUX NUMPY ---
arr_pil = np.array(img_pil, dtype=np.uint8)
arr_cv = np.array(img_cv_rgb, dtype=np.uint8)

# --- COMPARAISON NUMÉRIQUE ---
diff = np.abs(arr_pil.astype(int) - arr_cv.astype(int))
max_diff = diff.max()
mean_diff = diff.mean()

print(f"Différence max : {max_diff}")
print(f"Différence moyenne : {mean_diff:.4f}")
print("Identiques :", np.array_equal(arr_pil, arr_cv))

# --- AFFICHAGE DE QUELQUES PIXELS ---
h, w, _ = arr_pil.shape
sample_coords = [(h//2, w//2), (0,0), (h-1,w-1)]
print("\nExemples de pixels :")
for (y, x) in sample_coords:
    print(f"({y}, {x}) -> PIL: {arr_pil[y,x]}, CV2: {arr_cv[y,x]}")

```
